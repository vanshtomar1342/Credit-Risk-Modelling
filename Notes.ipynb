## THEORY

#### In banking we have two terminologies:

#### Asset: An asset is any product or service that can yield profit for the organisation. It includes housing loan, personal loan, education loan, vehicle loan, credit card, etc.

#### Liability: A liability is any product or service that does NOT yield any profit in the end. It includes current account, savings account, fixed deposits, recurring deposits, etc.

#### NOTE: An asset for bank is a liability for customer and an asset for customer is a liability for the bank

### NPA
#### NPA stands for Non-Performing Asset. It is a loan that gets defaulted, that is, the customer does not pay back the loan amount.

#### 1. Disbursed Amount: It is the loan amount given to a customer.
#### 2. OSP (Outstanding Principle): It is the remaining amount with customer that has to be paid to the bank by the customer at any point in time. As a customer pays EMIs, the OSP keeps decreasing. OSP should be zero at the end of the loan cycle.
#### 3. DPD (Days Past Due): It is the number of days past the due date of payment of EMI when the customer has yet not paid the EMI. DPD should be ideally zero. When DPD > 0, then it is said that the customer has "defaulted".
#### 4. PAR (Portfolio At Risk): It is the OSP when DPD > 0 days.
#### 5. NPA (Non-Performing Asset): It is a loan account where DPD > 90 days.

### Credit Risk Types In Banking

#### When DPD is 0 --> NDA (Non-Deliquint Account)
#### When DPD is between 0 to 30 --> SMA1 (Standard Monitoring Account)
#### When DPD is between 31 to 60 --> SMA2 (Standard Monitoring Account)
#### When DPD is between 61 to 90 --> SMA3 (Standard Monitoring Account)
#### When DPD is between 90 to 180 --> NPA (Non-Performing Asset)
#### When DPD is more than 180 --> Written-off (Loan which is not present)

### Q. Why are NPAs written-off after a DPD of 180 days?
### A. If NPAs are removed, then the loan portfolio quality of the bank will be better, which in turns makes the market sentiment better and finally it results in improvement of the stock price of the bank.

#### NPA is of two types:
#### 1. GNPA (Gross NPA) - It is usually 3% to 5% of the OSP which gets defaulted. If a bank has an OSP of Rs. 100 and the GNPA is 4%, it means that Rs. 4 will be defaulted and cannot be recovered by the bank.
#### 2. NNPA (Net NPA) - It is usually 0.01% to 0.06% of the OSP. It is the provisioning amount which is subtracted from the GNPA.

#### NOTE: In order to assess the quality of a bank, we should look at the GNPA value and not the NNPA. GNPA should be low for a good bank.
