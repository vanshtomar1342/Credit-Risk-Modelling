### HYPOTHESIS TESTING is a part of Inferential Statistics. It is used to draw inferences from the data.

### It involves the following steps:
####
#### 1. Declaring the Null Hypothesis (H0)
####
#### 2. Declaring the Alternate Hypothesis (Ha or H1): It is the opposite of null hypothesis
####
#### 3. Assuming a value of Alpha
#### Alpha is the significance level or strictness level. It is also known as margin of error. In technical terms, the significance level, also denoted as alpha or Î±, is the probability of rejecting the null hypothesis when it is infact true.
#### Low risk projects -> Less strictness on margin of error -> High Alpha
#### High risk projects -> More strictness on margin of error -> Low Alpha
####
#### 4. Determining the confidence interval
#### Confidence Interval = 1 - Alpha
#### In technical terms, confidence intervals measure the degree of uncertainty or certainty in a sampling method.
####
#### 5. Calculating the evidence against H0
#### This evidence is nothing but the p-value. The p value is a number, calculated from a statistical test, that describes how likely you are to have found a particular set of observations if the null hypothesis were true.
#### The p-value is calculated using several tests like Z-test, T-test, Chi-Square Test, ANOVA, etc. along with degree of freedom
####
#### 6. If p-value <= Alpha, then we reject H0
#### If p-value > Alpha, then we fail to reject H0
####
### Q. Why do we say "we fail to reject H0" instead of "we accept H0"?
### A. We say so because we never try to draw evidence in the favour of H0, instead we try to draw evidence against H0 and when we faill to draw such evidence, we say we have failed to reject H0.

### Q. Where to use which test?
### A. Chi-Square Test is used in case of categorical vs categorical variables
### T-test is used in case of categorical vs numerical variables when there are two categories in the categorical vairable
### ANOVA is used in case of categorical vs numerical variables when there are three or more categories in the categorical vairable

### Multicollinearity vs Correlation
#### Multicollinearity = Predictibility of each feature by other features
#### Correlation = Strength and nature of linear relationship between two features
####
#### NOTE: Correlation is specific to strictly linear relationship between columns/features. In case of non-linear relationship between features, correlation gives misleading values


#### In banking, the threshold VIF value is taken as 6 but it depends on the use case

#### There are two methods of VIF: Parallel and Sequential
#### Parallel VIF: Checking every feature with all other features iteratively
#### Sequential VIF: Checking every feature with all other features and dropping the features with high VIF (VIF > threshold) in each iteration
####
#### NOTE: Sequential VIF is the correct method but parallel VIF is a wrong method because in parallel VIF all the multicollinearly associated features will be dropped however we want to preserve atleast one of the unique features from any multicollinear association
